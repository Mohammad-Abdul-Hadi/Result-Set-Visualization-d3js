```python
import csv
```

```python
with open('data for emp study.csv', encoding="utf8") as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    out = "{\n\t\"name\": \"data\",\n\t\"children\": [{\n"
    for row in csv_reader:
        if line_count in range(1, 76):
            lc = line_count % 15
            counter = line_count % 15
            if lc in range(0, 15):
                if lc == 1:
                    out += "\t\t\"name\": \""+row[0]+"\",\n"
                    out += "\t\t\"children\": [{\n"
                # out += str(lc)+"\n"
                out += "\t\t\t\"name\": \""+row[1]+"\",\n\
                \t\"children\": [{\n"

                out += "\t\t\t\t\"name\": \"Accuracy: "+row[2]+"\",\n\
                \t\t\"size\": "+ str(float(row[2])*10000)+"\n\
                \t}, {\n"

                out += "\t\t\t\t\"name\": \"F1-score: "+row[3]+"\",\n\
                \t\t\"size\": "+ str(float(row[3])*10000)+"\n\
                \t}, {\n"

                out += "\t\t\t\t\"name\": \"Run-time: "+"5000"+" ms\",\n\
                \t\t\"size\": "+ "5000" +"\n\
                \t}]\n"
                
                if lc == 0:
                    out += "\t\t}]\n\t},{\n"
                else:
                    out += "\t\t\t},\n\t\t\t{\n"
        
        elif line_count in range(76, 91):
            lc = line_count % 3
            counter = line_count % 3
            if lc in range(0, 3):
                if lc == 1:
                    out += "\t\t\"name\": \""+row[0]+"\",\n"
                    out += "\t\t\"children\": [{\n"
                # out += str(lc)+"\n"
                out += "\t\t\t\"name\": \""+row[1]+"\",\n\
                \t\"children\": [{\n"

                out += "\t\t\t\t\"name\": \"Accuracy: "+row[2]+"\",\n\
                \t\t\"size\": "+ str(float(row[2])*10000)+"\n\
                \t}, {\n"

                out += "\t\t\t\t\"name\": \"F1-score: "+row[3]+"\",\n\
                \t\t\"size\": "+ str(float(row[3])*10000)+"\n\
                \t}, {\n"

                out += "\t\t\t\t\"name\": \"Run-time: "+"5000"+" ms\",\n\
                \t\t\"size\": "+ "5000" +"\n\
                \t}]\n"
                
                if lc == 0:
                    out += "\t\t}]\n\t},{\n"
                else:
                    out += "\t\t\t},\n\t\t\t{\n"
        
        line_count += 1
    
    out = out[:-5]
    out += "\t}]\n}"
    print(out)
    print(f'Processed {line_count} lines.')
```

```python
text_file = open("C:/xampp/htdocs/d3js_tree/data.json", "w")
n = text_file.write(out)
text_file.close()
```
